```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('vgsales.csv')

# Drop rows with missing values in sales columns
data = data.dropna(subset=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'])

# Density Plot for Sales by Region
plt.figure(figsize=(12, 6))

# Apply specific colors to each region's sales density plot
sns.kdeplot(data=data['NA_Sales'], label='North America', shade=True, color='#1a9850') 
sns.kdeplot(data=data['EU_Sales'], label='Europe', shade=True, color='#4575b4') 
sns.kdeplot(data=data['JP_Sales'], label='Japan', shade=True, color='#d73027')  
sns.kdeplot(data=data['Other_Sales'], label='Rest of World', shade=True, color='#806b71')  

# Set the x-axis to a logarithmic scale
plt.xscale('log')
plt.title('Density Plot of Video Game Sales by Region')
plt.xlabel('Sales (in millions)')
plt.ylabel('Density')
plt.legend()
plt.show()
```
