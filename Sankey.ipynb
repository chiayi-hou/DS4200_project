```python
import sankey as sk
import pandas as pd
```

```python
df_games = pd.read_csv("vgsales.csv")
df_games.head()
```

```python
df_grouped = df_games.groupby('Publisher')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales']].sum().reset_index()

# Sort the DataFrame by the 'Score' column in descending order
sorted_df = df_grouped.sort_values(by='Global_Sales', ascending=False)

# Select the top 100 rows
top_50 = sorted_df.head(50)

top_50[['NA_Sales', 'EU_Sales', 'JP_Sales']] *= 100

melted_df = pd.melt(top_50, id_vars=['Publisher'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales'],
                     var_name='Region', value_name='Sales')
melted_df.head()
```

```python
col = ["Publisher", "Region"]
fig = sk.make_sankey(melted_df, col, vals="Sales")

fig.update_layout(
    width=800,  # Set the width of the figure
    height=1200  # Set the height of the figure
)

fig.update_layout(title_text="Top 50 Platforms vs. Region")
fig.show()
```
